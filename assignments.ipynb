```python
def count(list1):
# Converting list to upper case
    l = [x.upper() for x in list1]
# create a empty dict [2]   
    dict1 = {}
#loop over the list
    for i in l:
#check if i is "in" items if true increment [1] 
        if i in dict1:
            dict1[i] +=1
        else:
            dict1[i] = 1
    return dict1
```

Python function called dicerolls that simulates rolling dice. 
This function will take 3 dice and roll then n-1 times to get pseduorandom values.
Create a list and adds the values to the list. 
The method count1 will count the occurances of the numbers.

```python
#k = the number of the dice. There are k dice
#n = the number of times the dice is rolled
def diceroll(k,n):
    list1 = []
        #roll the dice
    for _ in range(n-1):#for x in range(6)
        random_integer = random.randint(1, 6*k)
       # print(random_integer)
        list1.append(random_integer)
    return list1
print(count1(diceroll(3,1000)))
#print(diceroll(3,10))
```

```python
## these lines of code keep track of whether random.binomial returns values that are greater than 50.
x = random.binomial(n=100, p=0.5, size=1000)
#create a list
list = []
#loop over the list 
for n in x:
#check if the value at index n is greater than 50
    if(n >= 50):
        list.append("Heads")
    else:
        list.append("tails")
#call the count funtion
count(list)
```

A common example of Simpson's paradox involves the batting averages of players in professional baseball. 
It is possible for one player to have a higher batting average than another player each year for a number of years, but to have a lower batting average across all of those years. This phenomenon can occur when there are large differences in the number of at bats between the years. Mathematician Ken Ross[18] demonstrated this using the batting average of two baseball players, Derek Jeter and David Justice, during the years 1995 and 1996:[19]
