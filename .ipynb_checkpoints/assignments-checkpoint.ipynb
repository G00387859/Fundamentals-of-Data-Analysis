```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def count(list1):
# Converting list to upper case
    l = [x.upper() for x in list1]
# create a empty dict [2]   
    items = {}
#loop over the list
    for i in l:
#check if i is "in" items if true increment [1] 
        if i in items:
            items[i] +=1
        else:
            items[i] = 1
    return items
```

```python
#k = the number of the dice. There are k dice
#n = the number of times the dice is rolled
def diceroll(k,n):
    list1 = []
        #roll the dice
    for _ in range(n-1):#for x in range(6)
        random_integer = random.randint(1, 6*k)
       # print(random_integer)
        list1.append(random_integer)
    return list1
print(count1(diceroll(3,1000)))
#print(diceroll(3,10))
```

```python
## these lines of code keep track of whether random.binomial returns values that are greater than 50.
x = random.binomial(n=100, p=0.5, size=1000)
#create a list
list = []
#loop over the list 
for n in x:
#check if the value at index n is greater than 50
    if(n >= 50):
        list.append("Heads")
    else:
        list.append("tails")
#call the count funtion
count(list)
```

The below shows the bell shaped curve of the binomal distribuation. The probability of tossing a head or a tail is 0.5 or 50%.
The game is played 1000 times. The result show that the probability is 50%.

```python

#sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=True, kde=False)
plt.figure(figsize=(15,8))
#sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='binomial \nP value = 0.5')

sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=True, label='binomial \nP value = 0.5')
#sns.distplot(random.binomial(n=100, p=0.5, size=1000, histplot=True, label='binomial \nP value = 0.5')
plt.xlabel('Number of Trails/Times tossed')
plt.ylabel('p values')
plt.title('Game of Head or Tails')
plt.legend()
plt.show()
```
